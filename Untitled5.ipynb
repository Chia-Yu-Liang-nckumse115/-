```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# 讀取圖片（請先把牛排橫切面圖片命名為 steak.jpg 放在同一目錄）
img_path = "/content/test_image.JPG"  # 將路徑更新為 Colab 環境可用的路徑
img = Image.open(img_path).convert("RGB")

# 可選：縮小一點加快運算（例如長邊縮到 600 px）
max_size = 600
w, h = img.size
scale = min(max_size / max(w, h), 1.0)
if scale < 1.0:
    img = img.resize((int(w * scale), int(h * scale)))

img_np = np.array(img).astype(np.float32)

# 正規化到 0~1
rgb = img_np / 255.0
R = rgb[:, :, 0]
G = rgb[:, :, 1]
B = rgb[:, :, 2]

# 為了讓規則比較好寫，先算一些指標
brightness = (R + G + B) / 3.0      # 亮度
red_dominance = R - (G + B) / 2.0   # 紅色相對優勢
saturation = np.max(rgb, axis=2) - np.min(rgb, axis=2)  # 粗略飽和度

# 建一張熟度分類圖：0=背景/其他, 1=偏生, 2=五分, 3=七分以上
doneness_map = np.zeros(R.shape, dtype=np.int32)

# 這裡是簡單的 heuristic，你之後可以自己調參：
# 規則只是先給你一版可以跑的 demo

# 1. 偏生 (Rare): 顏色深、紅色明顯、飽和度高
cond_rare = (
    (red_dominance > 0.15) &
    (brightness < 0.55) &
    (saturation > 0.15)
)

# 2. 五分熟 (Medium-rare): 粉紅、亮度中等、飽和度適中
cond_med_rare = (
    (red_dominance > 0.05) &
    (brightness >= 0.55) & (brightness <= 0.75) &
    (saturation > 0.08)
)

# 3. 七分以上 (Medium+/Well-done): 偏灰棕，亮度較高或較低但紅色不那麼突出
cond_well = (
    (~cond_rare) & (~cond_med_rare) &
    (brightness > 0.45) &
    (R > 0.2) & (G > 0.15) & (B > 0.1)
)

doneness_map[cond_rare] = 1
doneness_map[cond_med_rare] = 2
doneness_map[cond_well] = 3

# 做一張可視化用的顏色圖：藍=生、綠=五分、紅=全熟
vis = np.zeros((*doneness_map.shape, 3), dtype=np.float32)

# Rare -> 藍色
vis[doneness_map == 1] = np.array([0.2, 0.3, 1.0])
# Medium-rare -> 綠色
vis[doneness_map == 2] = np.array([0.2, 0.9, 0.2])
# Well-done -> 紅色
vis[doneness_map == 3] = np.array([1.0, 0.2, 0.2])

# 讓沒被分類的地方用原圖顏色（背景 or 盤子等）
mask_unclassified = (doneness_map == 0)
vis[mask_unclassified] = rgb[mask_unclassified]

# 統計各熟度比例（只算被分類到 1~3 的像素）
total_pixels = np.sum(doneness_map > 0)
rare_ratio = np.sum(doneness_map == 1) / total_pixels if total_pixels > 0 else 0
med_rare_ratio = np.sum(doneness_map == 2) / total_pixels if total_pixels > 0 else 0
well_ratio = np.sum(doneness_map == 3) / total_pixels if total_pixels > 0 else 0

print("=== 熟度比例估計（僅供 demo，非精準科學計算）===")
print(f"偏生 (Rare)：{rare_ratio * 100:.1f}%")
print(f"五分熟 (Medium-rare)：{med_rare_ratio * 100:.1f}%")
print(f"七分以上 (Medium+/Well-done)：{well_ratio * 100:.1f}%")

# 顯示原圖與分類圖
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.title("原始牛排橫切面圖片")
plt.imshow(img_np.astype(np.uint8))
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("熟度分類視覺化（藍=生 綠=五分 紅=七分以上）")
plt.imshow(vis)
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
# ============================================
# 牛排：原圖 → Jet 光譜圖 → 由 Jet 反推 + 紅色強化
# - 支援上傳圖片 (Gradio 介面)
# - 專門把「偏紅的地方」變更紅，用來區分生與三分熟
# ============================================

import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import gradio as gr

# ------------------------------------------------------------
# 0. 工具函式：紅色選擇性增強（讓紅的地方更紅）
# ------------------------------------------------------------
def boost_red_selective(img_rgb, red_gain=1.4, v_gain=1.1):
    """
    img_rgb: uint8, RGB, [0,255]
    讓「紅色區域」的飽和度與亮度稍微拉高，
    達到生肉更紅、熟肉更偏褐的效果。
    """
    img = img_rgb.astype(np.uint8)

    # 轉 HSV（RGB → HSV）
    hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV).astype(np.float32)
    H, S, V = hsv[:,:,0], hsv[:,:,1], hsv[:,:,2]

    # 紅色範圍：H 在 [0,10] 或 [160,179]，且飽和度稍微高一點
    red_mask = ((H < 10) | (H > 160)) & (S > 40)

    # 增強紅色區域
    S[red_mask] = np.clip(S[red_mask] * red_gain, 0, 255)
    V[red_mask] = np.clip(V[red_mask] * v_gain, 0, 255)

    hsv[:,:,1], hsv[:,:,2] = S, V
    boosted = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB)
    return boosted


# ------------------------------------------------------------
# 1. 從原圖計算「熟度指標 D」並產生 Jet 光譜圖
# ------------------------------------------------------------
def image_to_jet(orig_rgb):
    """
    orig_rgb: (H,W,3) uint8, RGB
    回傳：
      D_norm: 0~1 的熟度指標
      jet_rgb: Jet 光譜圖 (uint8, RGB)
    """
    # 轉 float [0,1]
    x = orig_rgb.astype(np.float32) / 255.0
    R, G, B = x[:,:,0], x[:,:,1], x[:,:,2]

    # 一個簡單的「偏紅指標 vs 亮度」：
    # 生肉：R 高，G/B 相對低 → redness 高
    # 熟肉：變褐，R 不一定特別高，亮度也可能變
    redness = R - 0.5 * (G + B)
    redness = np.clip(redness, 0, None)

    # 同時考慮亮度，避免太暗的陰影被當成生肉
    brightness = 0.3*R + 0.59*G + 0.11*B

    # 熟度指標 D：越生 → 值越小，越熟 → 值越大
    # 這裡用 1 - redness + brightness 做一個簡單綜合
    D = (1.0 - redness) + 0.8 * brightness

    # Normalize 到 [0,1]
    D_min, D_max = D.min(), D.max()
    D_norm = (D - D_min) / (D_max - D_min + 1e-8)

    # 用 Jet colormap 映射
    cmap = cm.get_cmap('jet')
    jet = cmap(D_norm)[:, :, :3]  # 忽略 alpha
    jet_rgb = (jet * 255).astype(np.uint8)

    return D_norm, jet_rgb


# ------------------------------------------------------------
# 2. 從 Jet 光譜圖反推 D，再生成一張「牛排風格」還原圖
# ------------------------------------------------------------
def jet_to_steak_like(jet_rgb, blend_with=None, alpha=0.5):
    """
    jet_rgb: Jet 光譜圖 (H,W,3), uint8, RGB
    blend_with: 若提供原圖，可做顏色混合 (原圖 vs 合成牛排色)
    alpha: 合成圖權重（越大合成感越重）

    回傳：
      recon_rgb: 還原圖 (加了牛排色調 + 可選擇跟原圖混合)
    """
    H, W, _ = jet_rgb.shape

    # 建立 Jet 的 lookup table：256 階
    cmap = cm.get_cmap('jet')
    lut = cmap(np.linspace(0, 1, 256))[:, :3]  # (256,3), float

    # 將輸入 Jet 圖攤平成 (N,3)
    jet_flat = (jet_rgb.astype(np.float32) / 255.0).reshape(-1, 3)  # (N,3)

    # 計算與 lut 每一階的距離，找最近的 idx
    # N 可能不小，不過一般 512x512 也還能接受
    dist = ((jet_flat[:, None, :] - lut[None, :, :]) ** 2).sum(axis=2)  # (N,256)
    idx = np.argmin(dist, axis=1).astype(np.float32)  # 0~255

    # 反推 D_norm ≈ idx / 255
    D_rec = (idx / 255.0).reshape(H, W)

    # 把熟度指標 D_rec 映射到「牛排顏色梯度」
    # 0.0 → 深紅（生）
    # 0.5 → 粉紅帶點褐
    # 1.0 → 褐色（全熟表皮）

    # 定義幾個關鍵顏色（RGB, 0~1）
    deep_red   = np.array([0.7, 0.08, 0.1])   # 生的中心
    pinky_red  = np.array([0.9, 0.45, 0.5])   # 三分熟～五分熟
    brown_edge = np.array([0.55, 0.35, 0.20]) # 外圍褐色

    # 分兩段線性插值
    A = np.clip(D_rec * 2.0, 0, 1)          # 0~0.5 → 0~1
    B = np.clip((D_rec - 0.5) * 2.0, 0, 1)  # 0.5~1 → 0~1

    mid = deep_red[None,None,:]  * (1 - A[:,:,None]) + \
          pinky_red[None,None,:] * A[:,:,None]

    high = pinky_red[None,None,:] * (1 - B[:,:,None]) + \
           brown_edge[None,None,:] * B[:,:,None]

    steak_color = np.where(D_rec[:,:,None] < 0.5, mid, high)
    steak_rgb = np.clip(steak_color * 255.0, 0, 255).astype(np.uint8)

    # 如果提供原圖，就做混合，讓邊界與紋理更像照片
    if blend_with is not None:
        base = blend_with.astype(np.uint8)
        base = cv2.resize(base, (W, H))
        recon_rgb = cv2.addWeighted(steak_rgb, alpha, base, 1 - alpha, 0)
    else:
        recon_rgb = steak_rgb

    # 最後再給還原圖加一次「紅色增強」，讓生肉區域更突出
    recon_rgb = boost_red_selective(recon_rgb, red_gain=1.3, v_gain=1.05)

    return recon_rgb


# ------------------------------------------------------------
# 3. Pipeline：給一張圖，輸出 (原圖強化紅色, Jet, 還原圖)
# ------------------------------------------------------------
def steak_spectrum_pipeline(img):
    """
    給 Gradio 的主函式。
    img: PIL Image 或 numpy 陣列 (H,W,3), RGB
    """
    # 轉 numpy RGB uint8
    if isinstance(img, Image.Image):
        orig = np.array(img.convert("RGB"))
    else:
        orig = img[..., :3].astype(np.uint8)

    # Step 0: 先對原圖做「紅色強化」，讓生肉 vs 三分熟更清楚
    orig_boosted = boost_red_selective(orig, red_gain=1.35, v_gain=1.08)

    # Step 1: 原圖 → Jet 光譜圖
    D_norm, jet_rgb = image_to_jet(orig_boosted)

    # Step 2: Jet 光譜圖 → 反推熟度 → 合成牛排色調 + 與原圖混合
    recon_rgb = jet_to_steak_like(jet_rgb, blend_with=orig_boosted, alpha=0.6)

    return orig_boosted, jet_rgb, recon_rgb


# ------------------------------------------------------------
# 4. Gradio 介面：上傳牛排 → 顯示三張圖
# ------------------------------------------------------------
def main():
    demo = gr.Interface(
        fn=steak_spectrum_pipeline,
        inputs=gr.Image(type="pil", label="上傳牛排橫切面照片（RGB）"),
        outputs=[
            gr.Image(type="numpy", label="原圖（紅色強化後）"),
            gr.Image(type="numpy", label="Jet 光譜圖"),
            gr.Image(type="numpy", label="由 Jet 反推的牛排圖（紅色強化）"),
        ],
        title="Steak Doneness Spectrum Viewer",
        description="上傳一張牛排橫切面照片，系統會產生 Jet 光譜圖並嘗試從光譜反推一張「牛排風格還原圖」，同時強化偏紅區域，方便分辨生與三分熟。"
    )
    demo.launch(debug=True)


if __name__ == "__main__":
    main()
```

```python
!pip install gradio -q

import numpy as np
import cv2
from PIL import Image
import matplotlib.cm as cm
import gradio as gr

def steak_pipeline_from_pil(
    pil_img,
    gamma=0.9,
    w_temp=0.7,
    alpha_base=0.45,
    alpha_max=0.8,
):
    """
    輸入：
      pil_img   : 上傳圖片（PIL Image）
      gamma     : temp_smooth 的 gamma（對比強化程度）
      w_temp    : D = w_temp * D_temp + (1 - w_temp) * D_label
      alpha_base: 與原圖混合的基礎權重（越大越相信模型色）
      alpha_max : alpha_map 的上限（避免模型色太兇）

    輸出：
      四張 RGB uint8 圖片：
        - 原圖
        - 三色熟度分類圖
        - Jet 光譜圖
        - 還原影像
    """
    # 轉成 RGB + [0,1] numpy
    img_rgb = pil_img.convert("RGB")
    orig_np = np.array(img_rgb) / 255.0
    H, W, _ = orig_np.shape

    R, G, B = orig_np[..., 0], orig_np[..., 1], orig_np[..., 2]

    # --------------------------------------------------------
    # 1. 原圖 → 溫度代理值 temp_proxy
    # --------------------------------------------------------
    temp_proxy = 0.9 * R + 0.25 * G - 0.05 * B
    t_min, t_max = temp_proxy.min(), temp_proxy.max()
    temp_norm = (temp_proxy - t_min) / (t_max - t_min + 1e-8)

    # 簡單做一個三段熟度分類：1=生, 2=五分, 3=熟
    doneness_map = np.zeros((H, W), dtype=np.uint8)
    doneness_map[temp_norm < 0.33] = 1
    doneness_map[(temp_norm >= 0.33) & (temp_norm < 0.66)] = 2
    doneness_map[temp_norm >= 0.66] = 3

    # 三色熟度視覺化
    vis = np.zeros_like(orig_np)
    color1 = np.array([200,  80, 120]) / 255.0   # 生
    color2 = np.array([240, 160, 120]) / 255.0   # 五分
    color3 = np.array([ 90,  50,  40]) / 255.0   # 熟 / 外層
    vis[doneness_map == 1] = color1
    vis[doneness_map == 2] = color2
    vis[doneness_map == 3] = color3

    # --------------------------------------------------------
    # 2. temp_norm → 平滑 + gamma → D_temp
    # --------------------------------------------------------
    temp_smooth = cv2.GaussianBlur(temp_norm.astype(np.float32), (9, 9), 1.6)
    # gamma 可調：0.5~1.5，越小對比越強，越大越平順
    D_temp = np.power(np.clip(temp_smooth, 0, 1), gamma)

    # --------------------------------------------------------
    # 3. 分類基準熟度 D_label，與 D_temp 混合
    # --------------------------------------------------------
    D_label = np.zeros((H, W), dtype=np.float32)
    D_label[doneness_map == 1] = 0.25   # 生
    D_label[doneness_map == 2] = 0.55   # 五分
    D_label[doneness_map == 3] = 0.85   # 熟 / 外層（稍微保守一點）

    # w_temp 是 D_temp 的權重；(1 - w_temp) 是 D_label 的權重
    w_temp = float(w_temp)
    w_temp = np.clip(w_temp, 0.0, 1.0)
    D = w_temp * D_temp + (1.0 - w_temp) * D_label
    D = np.clip(D, 0, 1)

    # --------------------------------------------------------
    # 4. Jet 光譜圖
    # --------------------------------------------------------
    cmap = cm.get_cmap('jet')
    thermal = cmap(D)[..., :3]   # [0,1]

    # --------------------------------------------------------
    # 5. D → 還原牛排顏色（三段 mapping）
    # --------------------------------------------------------
    C_raw  = np.array([185,  40,  60]) / 255.0   # 深紅（偏生）
    C_med  = np.array([230, 120, 130]) / 255.0   # 粉紅（五分）
    C_well = np.array([200, 155, 130]) / 255.0   # 淺棕肉色（熟）
    C_sear = np.array([95,   60,  40]) / 255.0   # 深咖啡（焦外層）

    recon_color = np.zeros((H, W, 3), dtype=np.float32)

    # 區間 1：0~0.4 生→五分
    mask1 = D <= 0.4
    D1 = np.zeros_like(D)
    D1[mask1] = D[mask1] / 0.4
    recon_color[mask1] = ((1 - D1[mask1])[..., None] * C_raw
                          + D1[mask1][..., None] * C_med)

    # 區間 2：0.4~0.8 五分→熟
    mask2 = (D > 0.4) & (D <= 0.8)
    D2 = np.zeros_like(D)
    D2[mask2] = (D[mask2] - 0.4) / 0.4
    recon_color[mask2] = ((1 - D2[mask2])[..., None] * C_med
                          + D2[mask2][..., None] * C_well)

    # 區間 3：0.8~1 熟→焦色
    mask3 = D > 0.8
    D3 = np.zeros_like(D)
    D3[mask3] = (D[mask3] - 0.8) / 0.2
    recon_color[mask3] = ((1 - D3[mask3])[..., None] * C_well
                          + D3[mask3][..., None] * C_sear)

    # --------------------------------------------------------
    # 6. 紋理噪聲 + 與原圖混合（alpha_base / alpha_max 可調）
    # --------------------------------------------------------
    noise_level = 0.03
    noise = (np.random.rand(H, W, 3) - 0.5) * noise_level
    recon_noise = np.clip(recon_color + noise, 0, 1)

    img_norm = orig_np

    # alpha_base：基礎相信模型色的程度
    # alpha_max ：擋住太激進的情況（上限）
    alpha_base = float(alpha_base)
    alpha_max = float(alpha_max)
    alpha_base = np.clip(alpha_base, 0.0, 1.0)
    alpha_max = np.clip(alpha_max, 0.0, 1.0)
    if alpha_max < alpha_base:
        alpha_max = alpha_base  # 避免上限比底還低

    alpha_map = alpha_base + 0.2 * (D - 0.5)   # 中心線附近上下浮動
    alpha_map = np.clip(alpha_map, 0.3, alpha_max)[..., None]

    recon_blend = np.clip(alpha_map * recon_noise +
                          (1 - alpha_map) * img_norm, 0, 1)

    # --------------------------------------------------------
    # 7. 轉回 uint8 給 Gradio 顯示
    # --------------------------------------------------------
    orig_out    = (orig_np      * 255).astype(np.uint8)
    vis_out     = (vis          * 255).astype(np.uint8)
    thermal_out = (thermal      * 255).astype(np.uint8)
    recon_out   = (recon_blend  * 255).astype(np.uint8)

    return orig_out, vis_out, thermal_out, recon_out

# ============================================================
# Gradio 介面：上傳圖片 + 三組參數可調
# ============================================================
def gradio_wrapper(img, gamma, w_temp, alpha_base, alpha_max):
    pil_img = Image.fromarray(img.astype(np.uint8))
    return steak_pipeline_from_pil(
        pil_img,
        gamma=gamma,
        w_temp=w_temp,
        alpha_base=alpha_base,
        alpha_max=alpha_max,
    )

with gr.Blocks() as demo:
    gr.Markdown("## 🥩 牛排光譜調參實驗台\n上傳一張圖，調整 gamma / D 混合比例 / α 混合強度，找出你最喜歡的還原風格。")

    with gr.Row():
        inp = gr.Image(label="上傳牛排（或其他）圖片", type="numpy")

    with gr.Row():
        gamma_slider = gr.Slider(
            minimum=0.5, maximum=1.5, step=0.05, value=0.9,
            label="gamma（對比強化程度，越小越硬，越大越柔）"
        )
    with gr.Row():
        wtemp_slider = gr.Slider(
            minimum=0.0, maximum=1.0, step=0.05, value=0.7,
            label="w_temp：D = w_temp * D_temp + (1 - w_temp) * D_label"
        )
    with gr.Row():
        alpha_base_slider = gr.Slider(
            minimum=0.0, maximum=1.0, step=0.05, value=0.45,
            label="alpha_base：整體相信模型色的基礎程度"
        )
        alpha_max_slider = gr.Slider(
            minimum=0.3, maximum=1.0, step=0.05, value=0.8,
            label="alpha_max：最高相信模型色的上限"
        )

    with gr.Row():
        out1 = gr.Image(label="原始圖片")
        out2 = gr.Image(label="三色熟度分類圖")
        out3 = gr.Image(label="Jet 光譜圖")
        out4 = gr.Image(label="光譜還原牛排影像")

    btn = gr.Button("開始分析 / 更新結果")

    btn.click(
        fn=gradio_wrapper,
        inputs=[inp, gamma_slider, wtemp_slider, alpha_base_slider, alpha_max_slider],
        outputs=[out1, out2, out3, out4],
    )

demo.launch()
```
